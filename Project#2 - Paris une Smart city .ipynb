```python
hauteur_mean= df.groupby(['arrondissement'], as_index=False)['hauteur_m'].mean()
hauteur_mean
```

```python
arbre_count.columns
```

```python
# Renommer la colonne hauteur avant le merge 
hauteur_mean.rename(columns={'hauteur_m':'Hauteur_mean'})
```

```python
circonf_mean.rename(columns={'circonference_cm':'cinrconference_mean'})
```

```python
# Regrouper l'ensemble des variables dans un seul dataframe pour une meilleure visualisation et pour poursuivre la synthese.
merge_mean = pd.merge(arbre_count,hauteur_mean)
merge_mean = pd.merge(merge_mean, circonf_mean)
data_analysis = merge_mean.rename(columns={'genre':'count_arbres','hauteur_m':'Hauteur_mean','circonference_cm':'circonference_mean'})
data_analysis
```

```python
# Etablissement d'une regle de ponderation pour calculer le coefficient de contrainte das chaque arrondissement
# La formulation s'ecrit comme suit : 
# count_arbres_updated = count_arbres + (hauteur_mean / hauteur_max)*count_arbres + 
# (circonference_mean / circonference_max)*count_arbres.
```

```python
data_analysis['coeff_hauteur'] = data_analysis['Hauteur_mean']/35
data_analysis
```

```python
data_analysis['coeff_circonference'] = data_analysis['circonference_mean']/700
data_analysis
```

```python
data_analysis['count_arbres_updated'] = data_analysis['count_arbres']+data_analysis['count_arbres']*data_analysis['coeff_circonference']+data_analysis['count_arbres']*data_analysis['coeff_hauteur']
```

```python
data_analysis
```

```python
# nombre d'arbres au niveau de l'analyse a utiliser pour le calcul des ratio
data_analysis['count_arbres'].sum()
```

```python
# nombre d'arbres virtuel apres ponderation a utiliser dans le calcul des ratio
data_analysis['count_arbres_updated'].sum()
```

```python
# calcul des ratio par rapport au nombre initial
data_analysis['ratio_initial']=data_analysis['count_arbres']/200121
```

```python
# Calcul des ratio par rapport au nombre pondere
data_analysis['ratio_pondere']=data_analysis['count_arbres_updated']/282478
```
